# Visualizing Fair Market Rent Values.

Exploring fair market rent values calculated by the US Housing and Urban Development Department. This data is used to calculate guidelines for several federal programs. The actual values for rents vary greatly across the US. 

```
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

# fair market rent for a 2-bedroom apartment distribution

```
df = pd.read_csv('https://assets.datacamp.com/production/repositories/2210/datasets/a1fb97d60bfbcf0661e320a35a4615f4e8661a68/FY18_4050_FMRs.csv')
df.head()
```

```
sns.distplot(df.fmr_2)
plt.show()
```

```
sns.set()
df.fmr_2.plot.hist()
plt.show()
```

# distribution of 3 bedroom rents

```
sns.set(color_codes=True)
fig, ax = plt.subplots()
sns.distplot(df.fmr_3, color='m', ax=ax)
ax.set(xlabel="3 Bedroom Fair Market Rent")
plt.show()
```

# distribution of 1 bedroom rents

```
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=df['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)
ax.legend()
plt.show()
```

```
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
```

# Regression

```
sns.lmplot(data=df,x='pop2010',y='fmr_2')
sns.despine()
plt.show()
```
